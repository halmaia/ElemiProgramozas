```python
def collatzGenerator(y):
  while y != 1:
    y = y // 2 if y % 2 == 0 else 3 * y + 1
    yield y
```

```python
def getIntList(path):
  integers = []
  file = open(path)
  for row in file:
    integers.append(int(row))
  file.close()
  return integers


def getCount(lst, min_val, max_val):
  cnt = 0
  for v in lst:
    if v >= min_val and v < max_val:
      cnt += 1
  return cnt

lst = getIntList("/content/adatsor.txt")
print(getCount(lst, 161440, 183737))
```

```python
def readTempRad(path: str) -> tuple:
  with open(path) as file:
    next(file)
    temp, rad = [], []
    for row in file:
      vals = row.split(sep='\t')[1:]
      temp.append(float(vals[0]))
      rad.append(float(vals[1]))
    return (temp, rad)
```

```python
def avg(lst: list) -> float:
  return sum(lst) / len(lst)
```

```python
def stdDev(lst: list) -> float:
  c = 0.0
  lst_avg = avg(lst)

  for x in lst:
    d = x - lst_avg
    c += d * d

  c = c / (len(lst) - 1)
  return c ** 0.5
```

```python
def corr(x: list, y: list) -> float:
  x_avg = avg(x)
  y_avg = avg(y)

  x_std = stdDev(x)
  y_std = stdDev(y)

  lgt = len(x)

  c = 0.0

  for (x_i, y_i) in zip(x, y):
    c += (x_i - x_avg) * (y_i - y_avg)

  return c / (x_std * y_std * (lgt - 1))
```

```python
def filter(temp, rad):
  f_temp, f_rad = [], []
  for i in range(len(temp)):
    if rad[i] > 0.0:
      f_rad.append(rad[i])
      f_temp.append(temp[i])

  return f_temp, f_rad
```

```python
path = "/content/2021_09.txt"
temp, rad = readTempRad(path)
print(stdDev(temp))
f_temp, f_rad = filter(temp, rad)
print(corr(temp, rad))
print(corr(f_temp, f_rad))
```

## Időformázó

```python
def formatTimeString(n: int, unit: str) -> str:
  if n == 0: return ''
  ret = f"{n} {unit}"
  return ret if n == 1 else ret + 's'

def getHumanReadableTimeString(secs: int) -> str:
  if type(secs) is not int: "Unsupported input type! Use int only!" 
  if secs < 0: return "Secs should be positive or zero!"
  if secs == 0: return "Now"

  M = 60
  H = 60 * M
  D = 24 * H
  Y = 365 * D

  y, secs = divmod(secs, Y)
  d, secs = divmod(secs, D)
  h, secs = divmod(secs, H)
  m, secs = divmod(secs, M)

  l = []
  l.append(formatTimeString(y, 'year'))
  l.append(formatTimeString(d, 'day'))
  l.append(formatTimeString(h, 'hour'))
  l.append(formatTimeString(m, 'minute'))
  l.append(formatTimeString(secs, 'second'))
  
  s = ', '.join([v for v in l if v != ''])
  return s[::-1].replace(',', ' dna ', 1)[::-1]
```

```python
print(getHumanReadableTimeString(3600*24+60+3))
```

```python
def clamp(v):
  if v < 0:
    return 0
  if v > 255:
    return 255
  return v
```

```python
def getHistogram(img):

  nrows, ncols, bands = img.shape

  R = [0] * 256
  G = [0] * 256
  B = [0] * 256

  for r in range(nrows):
    for c in range(ncols):
      red, green, blue = img[r][c]
      R[red] += 1
      G[green] += 1
      B[blue] += 1
      

  plt.plot(R, c='red')
  plt.plot(G, c='green')
  plt.plot(B, c='blue')
  plt.show()
```

```python
img = openImage('/content/Dalia.jpg')
getHistogram(img)
```

```python
def rotateLeft(img): 
  nrows, ncols, bands = img.shape
  img_rot = plt.np.empty((ncols, nrows, bands), dtype=img.dtype)
  for r in range(nrows):
    for c in range(ncols):
      img_rot[c][r] = img[r][ncols-c-1]
  return img_rot
```

```python
def flip(img): 
  nrows, ncols, bands = img.shape
  img_flip = img.copy()
  for r in range(nrows):
    img_flip[r] = img[r][::-1]
  return img_flip
```
